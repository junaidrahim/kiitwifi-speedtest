```python
import json, matplotlib
from datetime import datetime

data = open("data.json", "r")

speed_dict = json.loads(data.read())

x = []
y_upload = []
y_download = []

for timestamp in [*speed_dict]:
    time_format = "%Y-%m-%d %H:%M:%S.%f"
    dt = datetime.strptime(timestamp, time_format)
    x.append(dt)

# download
for timestamp in [*speed_dict]:
    speeds = speed_dict[timestamp]
    y_download.append(float(speeds[0].split(" ")[1]))

# upload
for timestamp in [*speed_dict]:
    speeds = speed_dict[timestamp]
    y_upload.append(float(speeds[1].split(" ")[1]))


dates = matplotlib.dates.date2num(x)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Download speed chart
%matplotlib inline
fig, ax = plt.subplots(1, figsize=(15,10))

fig.autofmt_xdate()
xfmt = mdates.DateFormatter('%d-%m-%y %H:%M')
ax.xaxis.set_major_formatter(xfmt)

plt.xlabel("Date and Time")
plt.ylabel("Download Internet Speed on KIIT-Wifi in Mbits/s")

plt.plot(x,y_download)
```
